```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cost_function(x):
  return (x+3)**2
```

```python
def gradient(x):
  return 2*(x+3)
```

```python
learning_rate = 0.1
num_iterations = 50
x_start = 2
```

```python
x = x_start
trajectory = [x]
```

```python
for _ in range(num_iterations):
  x -= learning_rate * gradient(x)
  trajectory.append(x)
```

```python
x_values = np.linspace(-10, 4, 400)
y_values = cost_function(x_values)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='y = (x + 3)^2', color='b')
plt.scatter(trajectory, [cost_function(x) for x in trajectory], c='r', label='Gradient Descent Steps')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gradient Descent to Find Local Minimum')
plt.legend()
plt.grid(True)
plt.show()
```
